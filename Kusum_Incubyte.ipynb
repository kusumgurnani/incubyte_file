```python
class SC:
    def __init__(self, x, y, z, d):
        self.x = x
        self.y = y
        self.z = z
        self.d = d

    def Movef(self):
        if self.d == 'N':
            self.y += 1
        elif self.d == 'S':
            self.y -= 1
        elif self.d == 'E':
            self.x += 1
        elif self.d == 'W':
            self.x -= 1
        elif self.d == 'U':
            self.z += 1
        elif self.d == 'D':
            self.z -= 1

    def Moveb(self):
        if self.d == 'N':
            self.y -= 1
        elif self.d == 'S':
            self.y += 1
        elif self.d == 'E':
            self.x -= 1
        elif self.d == 'W':
            self.x += 1
        elif self.d == 'U':
            self.z -= 1
        elif self.d == 'D':
            self.z += 1

    def Turnl(self):
        if self.d == 'N':
            self.d = 'W'
        elif self.d == 'S':
            self.d = 'E'
        elif self.d == 'E':
            self.d = 'N'
        elif self.d == 'W':
            self.d = 'S'
        elif self.d == 'U':
            self.d = 'W'
        elif self.d == 'D':
            self.d = 'E'

    def Turnr(self):
        if self.d == 'N':
            self.d = 'E'
        elif self.d == 'S':
            self.d = 'W'
        elif self.d == 'E':
            self.d = 'S'
        elif self.d == 'W':
            self.d = 'N'
        elif self.d == 'U':
            self.d = 'E'
        elif self.d == 'D':
            self.d = 'W'

    def Turnup(self):
        if self.d not in ['D']:
            self.d = 'U'

    def Turndown(self):
        if self.d not in ['U']:
            self.d = 'D'

    def cmd(self, com):
        if com == 'f':
            self.Movef()
        elif com == 'b':
            self.Moveb()
        elif com == 'l':
            self.Turnl()
        elif com == 'r':
            self.Turnr()
        elif com == 'u':
            self.Turnup()
        elif com == 'd':
            self.Turndown()
```

```python
def cmd(x, y, z, d, com):
    obj = SC(x, y, z, d)

    for comm in com:
        obj.cmd(comm)

    return obj.x, obj.y, obj.z, obj.d
```

```python
x = 0
y = 0
z = 0
d = 'N'
com = []
n=int(input("Enter number of commands:"))
print("Directions Variable: \n UP : u \n Down : d \n Right : r \n Left : l \n Forward : f \n Backward : b \n")
for i in range(n):
    c=input("Enter direction variable using above commands:")
    c=c.lower()
    if c not in ['u','d','l','r','f','b']:
        print("Invalid Direction entered! \n Enter Direction again")
    else:
        com.append(c)

ansx, ansy, ansz, ansd = cmd(x, y, z, d, com)

print("Final Position:","(",ansx,",", ansy,",", ansz,")")
print("Final Direction:", ansd)
```
